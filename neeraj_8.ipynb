```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
data=np.array([0,1,1,0,1,1])
print("Data =",data)
```

```python
f=2
t=np.arange(0,1,1/500)
x=np.cos(2*pi*f*t)
c1=np.cos(2*pi*f*t+pi/4)
c2=np.cos(2*pi*f*t+3*pi/4)
c3=np.cos(2*pi*f*t+5*pi/4)
c4=np.cos(2*pi*f*t+7*pi/4)
```

```python
data_even=[data[i] for i in range(len(data)) if (i+1)%2==0]
data_odd=[data[i] for i in range(len(data)) if (i+1)%2==1]
print(data_even)
print(data_odd)
```

```python
data_even=np.repeat(data_even,2)
data_odd=np.repeat(data_odd,2)
print(data_even)
print(data_odd)
```

```python
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(10,10))
lw=3
fs=18
z=0
for i in range(len(data)-1):
  #carrier
  ax1.plot(t+z,x,linewidth=lw)
  ax1.set_ylabel("Carrier signal",fontsize=fs)
  ax1.grid(which='major',color='k',linestyle='--')
  #qpsk
  if data_odd[i+1]==1:
    if data_even[i]==1:
      ax2.plot(t+z,c4,linewidth=lw)
    else:
      ax2.plot(t+z,c3,linewidth=lw)
  else:
    if data_even[i]==1:
      ax2.plot(t+z,c1,linewidth=lw)
    else:
      ax2.plot(t+z,c2,linewidth=lw)
  ax2.set_ylabel("OFFSET QPSK",fontsize=fs)
  ax2.grid(which='major',color='k',linestyle='--')
  z=z+1
```

```python
ps=1
tb=1
f=np.linspace(-8,8,500)
f0=2
PSD=ps*tb*(np.sinc((f-f0)*2*tb))**2
PSD+=ps*tb*(np.sinc((f+f0)*2*tb))**2
plt.plot(f,PSD)
```
