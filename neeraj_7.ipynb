```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
data=np.array([1,0,1,1,0,0,1])
print("Data =",data)
```

```python
t=np.arange(0,1,1/500)
fl=2
fh=4
xl=np.sin(2*pi*fl*t)
xh=np.sin(2*pi*fh*t)
z=0
lw=3
fs=18
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(10,10))

for i in range(len(data)):
  data_psk=2*data-1

  ax1.plot(t+z,xl,linewidth=lw)
  ax1.set_ylabel("Carrier signal",fontsize=fs)
  ax1.grid(which='major',color='k',linestyle='--')

  ax2.plot(t+z,xl*data[i],linewidth=lw)
  ax2.set_ylabel("ASK signal",fontsize=fs)
  ax2.grid(which='major',color='k',linestyle='--')

  ax3.plot(t+z,xl*data_psk[i],linewidth=lw)
  ax3.set_ylabel("PSK signall",fontsize=fs)
  ax3.grid(which='major',color='k',linestyle='--')

  if data[i]==0:
    ax4.plot(t+z,xl,linewidth=lw)
  else:
    ax4.plot(t+z,xh,linewidth=lw)

  ax4.set_ylabel("FSK signall",fontsize=fs)
  ax4.grid(which='major',color='k',linestyle='--')
  z=z+1
```
