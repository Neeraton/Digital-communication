```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bit=np.array([1,0,1,0,1,0])
print(bit)
```

```python
bitrate=1
n=500
T=len(bit)/bitrate
N=n*len(bit)
dt=T/N
t=np.arange(0,T,dt)

u_nrz=np.zeros((N,1))
u_rz=np.zeros((N,1))
p_nrz=np.zeros((N,1))
p_rz=np.zeros((N,1))
```

```python
#u_nrz
for i in range(len(bit)-1):
  if bit[i]==1:
    u_nrz[i*n+1:(i+1)*n+1]=1
  else:
     u_nrz[i*n+1:(i+1)*n+1]=0
plt.plot(t,u_nrz)
plt.grid()
```

```python
#p_nrz
for i in range(len(bit)-1):
  if bit[i]==1:
    p_nrz[i*n+1:(i+1)*n+1]=1
  else:
     p_nrz[i*n+1:(i+1)*n+1]=-1
plt.plot(t,p_nrz)
plt.grid()
```

```python
#u_rz
for i in range(len(bit)-1):
  if bit[i]==1:
    u_rz[i*n+1:int((i+0.5)*n+1)]=1
    u_rz[int((i+0.5)*n+1):(i+1)*n+1]=0
  else:
     u_rz[i*n+1:(i+1)*n+1]=0
plt.plot(t,u_rz)
plt.grid()
```

```python
#p_rz
for i in range(len(bit)-1):
  if bit[i]==1:
    p_rz[i*n+1:int((i+0.5)*n+1)]=1
    p_rz[int((i+0.5)*n+1):(i+1)*n+1]=0
  else:
    p_rz[i*n+1:int((i+0.5)*n+1)]=-1
    p_rz[int((i+0.5)*n+1):(i+1)*n+1]=0
plt.plot(t,p_rz)
plt.grid()
```
